```python
raw_df['전력거래량(MWh)']=raw_df['전력거래량(MWh)'].astype(int)
```

```python
raw_df = raw_df[['연료원','거래일','전력거래량(MWh)']]
raw_df
```

```python
raw_df['거래일']=pd.to_datetime(raw_df['거래일'])
```

```python
raw_df.dtypes
```

```python
raw_df['Year']=raw_df['거래일'].dt.year
raw_df['Month']=raw_df['거래일'].dt.month
raw_df['Day']=raw_df['거래일'].dt.day
raw_df
```

```python
raw_df=raw_df[['연료원','전력거래량(MWh)','Year','Month','Day']]
raw_df
```
